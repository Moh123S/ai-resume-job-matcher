```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
resumes_path = '../data/resumes/'
jd_path = '../data/job_descriptions/'
```

```python
resume_files = os.listdir(resumes_path)
print("Total Resumes:", len(resume_files))
print("Resume Files Found:",resume_files)

resume_lengths = []
```

```python
import fitz
```

```python
for file in resume_files:
    if file.endswith(".pdf"):
        full_path = os.path.join(resumes_path,file)
        text = ""
        with fitz.open(full_path) as doc:
            for page in doc:
                text += page.get_text()
        word_count = len(text.split())
        print(f"{file} --> {word_count} words")
        resume_lengths.append(word_count)        
```

```python
if resume_lengths:
    plt.figure(figsize=(6,4))
    sns.histplot(resume_lengths,bins=5,kde=True,color='skyblue')
    plt.title("Resume Word Count Distribution")
    plt.xlabel("Number of Words")
    plt.ylabel("Frequency")
    plt.show()
else:
    print("No resume files found or empty text.")    
```

```python
jd_files = os.listdir(jd_path)
print("Total Job Descriptions:", len(jd_files))
```

```python
jd_lengths = [len(open(os.path.join(jd_path,f),'r', encoding='utf-8').read().split())
                  for f in jd_files if f.endswith('.txt')]
plt.figure(figsize=(10,5))
sns.histplot(jd_lengths,bins=30)
plt.title('job Length Distribution')
plt.xlabel('Number of Words')
plt.ylabel('Frequency')
plt.show()
```
